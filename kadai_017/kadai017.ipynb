```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name} は大人です。")
        else:
            print(f"{self.name} は大人ではありません。")

humans = [
    Human("一郎", 25),
    Human("花子", 18),
    Human("二郎", 20),
    Human("三郎", 15)
]

for human in humans:
    human.check_adult()
```
