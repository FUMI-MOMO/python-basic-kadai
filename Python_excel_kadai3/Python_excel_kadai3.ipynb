```python
import pandas as pd
import openpyxl
from openpyxl.styles import PatternFill,Font,Alignment

file_list = ['2022_年間売上表.xlsx','2023_年間売上表.xlsx']

df = pd.DataFrame()

for file_name in file_list:
    temp_df = pd.read_excel(file_name)
    temp_df.columns = ['年売上','商品','金額(千円)']
    df = pd.concat([df,temp_df],ignore_index=True)

grouped = df.groupby(['商品','年売上'])['金額(千円)'].sum()

writer = pd.ExcelWriter('売上集計表.xlsx')
grouped.to_excel(writer,sheet_name='売上集計表',index=True)
writer.close()

book = openpyxl.load_workbook('売上集計表.xlsx')
sheet = book['売上集計表']

gray_fill = PatternFill(patternType='solid',fgColor="F2F2F2")
for cell in sheet[1]:
    cell.fill = gray_fill
    cell.font = Font(bold=True)
    cell.alignment = Alignment(horizontal='center')

book.save('売上集計表.xlsx')
```
