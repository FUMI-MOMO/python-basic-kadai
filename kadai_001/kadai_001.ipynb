```python
import openpyxl
from datetime import datetime

today = datetime.today().strftime("%Y%m%d")

workbook = openpyxl.Workbook()
ws = workbook.active
ws.title = "請求書"

ws['B4'] = "株式会社株式ＡＢＣ"
ws['B5'] = "〒101-0022 東京都千代田区神田練堀町300"
ws['B6'] = "ＴＥＬ:03-1234-5678 ＦＡＸ:03-1234-5678"
ws['B7'] = "担当者名:鈴木一郎様"

ws['F4'] = "No."
ws['G4'] = "0001"
ws['F5'] = "日付"
ws['G5'] = "2026/01/26"

headers = ["商品名", "数量", "単価", "金額"]
start_row = 10

for col, header in enumerate(headers, start=2):
    cell = ws.cell(row=start_row, column=col)
    cell.value = header

items = [
    ["商品A", 2, 10000, 20000],
    ["商品B", 1, 15000, 15000],
]

for row_offset, item in enumerate(items, start=1):
    for col_offset, value in enumerate(item):
        cell = ws.cell(row=start_row + row_offset, column=2 + col_offset)
        cell.value = value

ws['E15'] = "=sum(E11:E12)"
ws['E16'] = "=E15*0.1"
ws['E17'] = "=E15+E16"


ws['B15'] = "小計"
ws['B16'] = "消費税"
ws['B17'] = "合計"

ws.column_dimensions["B"].width = 20
ws.column_dimensions["C"].width = 10
ws.column_dimensions["D"].width = 12
ws.column_dimensions["E"].width = 12
ws.column_dimensions["F"].width = 8
ws.column_dimensions["G"].width = 12

workbook.save(f"請求書_{today}.xlsx")
```
